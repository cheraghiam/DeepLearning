## 11. Lazy Evaluation â­

```python
lf = (
    pl.scan_csv("data.csv")
      .filter(pl.col("A") > 10)
      .groupby("category")
      .agg(pl.mean("B"))
)

lf.collect()   # execution happens here
```

Benefits:

* Predicate pushdown
* Projection pushdown
* Query optimization
