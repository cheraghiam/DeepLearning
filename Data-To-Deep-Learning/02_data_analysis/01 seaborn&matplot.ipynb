## üìà Line Plot

```python

```

# üß™ Practical Examples (NumPy + Pandas)

---

## Example 1Ô∏è‚É£: Numerical Distribution



````python
np.random.seed(42)

data = np.random.normal(loc=70, scale=10, size=500)
df = pd.DataFrame({'score': data})

sns.histplot(df['score'], bins=30, kde=True)
plt.title("Score Distribution")
plt.show()

sns.boxplot(x=df['score'])
plt.show()
````



‚úî Use case: Detect outliers and check normality

---

## Example 2Ô∏è‚É£: Group Comparison



````python

df = pd.DataFrame({
    'group': ['A','A','A','B','B','B'],
    'value': [10,12,11,20,22,19]
})

sns.barplot(x='group', y='value', data=df)
plt.show()

sns.boxplot(x='group', y='value', data=df)
plt.show()
````



‚úî Use case: Compare classes or sensor outputs

---

## Example 3Ô∏è‚É£: Scatter + Regression



````python
x = np.linspace(0, 10, 50)
y = 3*x + np.random.randn(50)*2

df = pd.DataFrame({'x': x, 'y': y})

sns.scatterplot(x='x', y='y', data=df)
sns.regplot(x='x', y='y', data=df)
plt.show()
````

‚úî Use case: Correlation analysis
---

## Example 4Ô∏è‚É£: Time Series / Signal Data



````python
time = pd.date_range('2024-01-01', periods=200, freq='ms')
signal = np.sin(2*np.pi*5*np.linspace(0,1,200)) + 0.2*np.random.randn(200)

df = pd.DataFrame({'time': time, 'signal': signal})

plt.plot(df['time'], df['signal'])
plt.title("Signal over Time")
plt.show()
````



‚úî Use case: ECG, EMG, EEG, sensor signals

---

## Example 5Ô∏è‚É£: Correlation Heatmap



````python
np.random.seed(0)
df = pd.DataFrame(np.random.randn(100,4), columns=['A','B','C','D'])

sns.heatmap(df.corr(), annot=True)
plt.show()
````



‚úî Use case: Feature selection

---

## Example 6Ô∏è‚É£: Fast Exploratory Data Analysis



````python
sns.pairplot(df)
plt.show()
````



‚úî Use case: Global dataset overview
